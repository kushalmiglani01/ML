```python
covid_germany = covid[covid.location == 'Germany']
```

```python
covid_italy = covid[covid.location == 'Italy']
```

```python
covid_italy_germany = covid_germany.merge(covid_italy, how='left', on='date', suffixes=('_germany', '_italy'))
```

```python
covid_italy_germany.shape
```

```python
covid_italy_germany.new_cases_germany.max()
```

```python
covid_italy_germany['total_cases_germany'] = covid_italy_germany.new_cases_germany.cumsum()
```

```python
covid_italy_germany['total_cases_italy'] = covid_italy_germany.new_cases_italy.cumsum()
```

```python
covid_italy_germany['total_cases_delta_ger_ital'] = covid_italy_germany.total_cases_germany - covid_italy_germany.total_cases_italy
```

```python
covid_italy_germany_fil = covid_italy_germany[(covid_italy_germany.date >= '2020-02-28') & (covid_italy_germany.date <= '2020-03-20')].copy()
```

```python
covid_italy_germany_fil['date'] = pd.to_datetime(covid_italy_germany_fil.loc[:,'date'])
covid_italy_germany_fil.head(2)
```

```python
covid_italy_germany_fil['days'] = (covid_italy_germany_fil['date'] - covid_italy_germany_fil['date'].min()).dt.days
```

```python
from scipy.optimize import curve_fit
import numpy as np

def expo_func(x, A, B):
  return A * np.exp(B*x)

params, covariance  = curve_fit(expo_func,
                                covid_italy_germany_fil.days,
                                covid_italy_germany_fil.total_cases_italy)
```

```python
params
```

```python
A, B = params
```

```python
predictions = expo_func(covid_italy_germany_fil.days, A, B)
```

```python
covid_italy_germany_fil['pred'] = predictions
```
