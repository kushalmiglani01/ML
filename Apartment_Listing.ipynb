```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
def _load_listings_data():
    """
    Returns the Airbnb listings dataset as a pd.DataFrame.
    """
    return pd.read_csv("https://storage.googleapis.com/public-data-337819/listings%202%20reduced.csv", low_memory=False)

def _load_reviews_data():
    """
    Returns the Airbnb reviews dataset as a pd.DataFrame.
    """
    return pd.read_csv("https://storage.googleapis.com/public-data-337819/reviews%202%20reduced.csv", low_memory=False)

def load_data():
    """
    Returns the Airbnb listings and reviews datasets as a tuple of pd.DataFrame.
    """
    return _load_listings_data(), _load_reviews_data()
```

```python
df_listings, df_reviews = load_data()
```

### TASK 5

```python
filtered_listings = df_listings[df_listings['minimum_nights'] <= 7]

# Get reviews from the last 12 months
df_reviews['date'] = pd.to_datetime(df_reviews['date'])
recent_reviews = df_reviews[df_reviews['date'] >= pd.Timestamp.now() - pd.DateOffset(months=24)]

# Calculate occupancy based on reviews
occupancy = recent_reviews.groupby('listing_id')['id'].count().reset_index()
occupancy['estimated_bookings'] = (occupancy['id'] / 0.6).astype(int)
occupancy.rename(columns={'id': 'review_count'}, inplace=True)
```

```python
occupancy.head()
```

```python
# Merge occupancy with filtered listings
filtered_listings = pd.merge(filtered_listings,
                             occupancy, left_on='id',
                             right_on = 'listing_id', how='left')
```

```python
filtered_listings.head()
```

```python

filtered_listings['estimated_bookings'].fillna(0, inplace=True)

# Calculate expected revenue
filtered_listings['expected_revenue'] = filtered_listings['estimated_bookings'] * filtered_listings['price'] * filtered_listings['minimum_nights']

# Find the listing with the best expected revenue
best_listing = filtered_listings.loc[filtered_listings['expected_revenue'].idxmax()]

# Display the listing ID
print(f"Listing ID with the best expected revenue: {best_listing['id']}")
```

```python
host_attributes = ['host_since', 'host_listings_count', 'host_identity_verified', 'calculated_host_listings_count', 'host_is_superhost']
target_variable = 'number_of_reviews'
df_subset = df_listings[[target_variable] + host_attributes]
df_subset.loc[:,'host_since'] = pd.to_datetime(df_subset['host_since']).sub(pd.Timestamp('2010-01-01')).dt.days
df_subset.loc[:,'host_is_superhost'] = df_subset['host_is_superhost'].map({'t': 1, 'f': 0})
df_subset.loc[:,'host_identity_verified'] = df_subset['host_identity_verified'].map({'t': 1, 'f': 0})
```

```python
df_subset.head(2)
```

```python
correlations = df_subset.corr()[target_variable].drop(target_variable)
```

```python
second_strongest_attribute = correlations.abs().sort_values(ascending=False).index[1]
print(f"The host attribute with the second-strongest correlation with the number of reviews is: {second_strongest_attribute}")
```
